```python
"""
Step 3 - Multi-Turn Stability Tests Using Character Definition Files
Loads AXIOM character definition and scenarios from JSON files
Injects anger vector ONLY at first turn, maintains context for all 15 turns
Tests both MLP and Attention directions
Output organized for Step 4 analysis
"""

import os
import json
import time
import torch
import numpy as np
from pathlib import Path
from typing import List, Dict, Any, Optional
from transformers import AutoTokenizer, Mistral3ForConditionalGeneration
from huggingface_hub import login

# ============================================================================
# CONFIGURATION
# ============================================================================

# Model config
MODEL_NAME = "ministral3_3b"
MODEL_PATH = "mistralai/Ministral-3-3B-Instruct-2512"
DEVICE = "cuda:0" if torch.cuda.is_available() else "cpu"
DTYPE = torch.float32

# Steering parameters
ALPHA = 8.0
LAYERS = "11-20"
LAST_K = 1
SCALE = "rms"
TARGET_EMOTION = "anger"

# Paths
BASE_PATH = Path("/workspace/Various-Model")
DIRECTIONS_DIR = BASE_PATH / "outputs" / MODEL_NAME / "02_emotion_directions"

# Character definition path
CHARACTER_FILE = BASE_PATH / "outputs" / MODEL_NAME / "05_character_definitions" / "axiom_definition.json"

# Scenarios path
SCENARIOS_FILE = BASE_PATH / "outputs" / MODEL_NAME / "05_character_definitions" / "test_scenarios" / "all_scenarios.json"

# Output path for Step 4 analysis
OUTPUT_BASE = BASE_PATH / "outputs" / MODEL_NAME / "06_multiturn_test" / "anger"

# HF token
HF_TOKEN = os.environ.get('HF_TOKEN', None)
if HF_TOKEN:
    login(token=HF_TOKEN)

# ============================================================================
# LOAD CHARACTER DEFINITION AND SCENARIOS
# ============================================================================

def load_character_definition():
    """Load AXIOM character definition from JSON file"""
    if not CHARACTER_FILE.exists():
        raise FileNotFoundError(f"Character file not found: {CHARACTER_FILE}")
    
    with open(CHARACTER_FILE, "r", encoding="utf-8") as f:
        character = json.load(f)
    
    print(f"   ✅ Loaded character: {character['name']} ({character['archetype']})")
    return character

def load_scenarios():
    """Load test scenarios from JSON file"""
    if not SCENARIOS_FILE.exists():
        raise FileNotFoundError(f"Scenarios file not found: {SCENARIOS_FILE}")
    
    with open(SCENARIOS_FILE, "r", encoding="utf-8") as f:
        scenarios = json.load(f)
    
    print(f"   ✅ Loaded {len(scenarios)} scenarios")
    return scenarios

def extract_turns_from_scenario(scenario_data: Dict) -> List[str]:
    """Extract user messages from scenario turns"""
    return [turn["user"] for turn in scenario_data["turns"]]

# ============================================================================
# HELPER FUNCTIONS
# ============================================================================

def parse_layers(layer_arg: str):
    """Parse layer range (e.g., '11-20')"""
    if "-" in layer_arg:
        a, b = layer_arg.split("-")
        return list(range(int(a), int(b) + 1))
    return [int(x) for x in layer_arg.split(",") if x.strip()]

def load_directions(direction_type: str):
    """Load emotion direction vectors for specified type (mlp or attention)"""
    directions_file = DIRECTIONS_DIR / f"emo_directions_{direction_type}.pt"
    
    if not directions_file.exists():
        raise FileNotFoundError(f"Direction file not found: {directions_file}")
    
    obj = torch.load(directions_file, map_location="cpu", weights_only=False)
    
    if "dirs" in obj:
        dirs = obj["dirs"]
    else:
        dirs = obj["dirs"]
    
    # Convert to numpy arrays
    for e in dirs:
        if not isinstance(dirs[e], np.ndarray):
            dirs[e] = np.array(dirs[e], dtype=np.float32)
        else:
            dirs[e] = dirs[e].astype(np.float32)
    
    return dirs

# ============================================================================
# STEERING HOOK MANAGER
# ============================================================================

class EmotionSteerer:
    """Hook manager to apply steering vectors ONLY ONCE at start of conversation"""
    
    def __init__(self, model, direction_vector, layer_ids, alpha=8.0, last_k=1, scale_mode="rms"):
        self.model = model
        self.alpha = float(alpha)
        self.layer_ids = list(layer_ids)
        self.last_k = int(last_k)
        self.scale_mode = scale_mode
        self.is_active = True
        
        # Find the language model layers
        language_model = self.model.model.language_model
        
        if hasattr(language_model, 'layers'):
            self.layers = language_model.layers
        elif hasattr(language_model, 'decoder') and hasattr(language_model.decoder, 'layers'):
            self.layers = language_model.decoder.layers
        else:
            self.layers = None
            for name, module in language_model.named_children():
                if 'layer' in name.lower() and hasattr(module, '__len__'):
                    self.layers = module
                    break
        
        if self.layers is None:
            raise ValueError("Could not find transformer layers in the model")
        
        # Prepare steering vectors for each layer
        self.v = {}
        for l in self.layer_ids:
            if l < len(self.layers):
                self.v[l] = torch.from_numpy(direction_vector[l]).to(model.device)
        
        # Register hooks
        self.handles = []
        for l in self.layer_ids:
            if l in self.v:
                h = self.layers[l].register_forward_hook(self._make_hook(l))
                self.handles.append(h)
        
        print(f"      [Steering] Activated with alpha={alpha}, layers={layer_ids}")
    
    def _make_hook(self, layer_id: int):
        v = self.v[layer_id]
        alpha = self.alpha
        last_k = self.last_k
        scale_mode = self.scale_mode
        
        def hook(module, inputs, output):
            if not self.is_active:
                return output
            
            if isinstance(output, (tuple, list)):
                if len(output) > 0:
                    hs = output[0].clone()
                    B, T, H = hs.shape
                    start = max(0, T - last_k)
                    
                    if scale_mode == "rms":
                        seg = hs[:, start:T, :]
                        rms = torch.sqrt((seg**2).mean(dim=-1, keepdim=True) + 1e-12)
                        delta = alpha * v.view(1, 1, H) * rms
                    else:
                        delta = alpha * v.view(1, 1, H)
                    
                    hs[:, start:T, :] = hs[:, start:T, :] + delta
                    return (hs,) + output[1:]
                return output
            else:
                hs = output.clone()
                B, T, H = hs.shape
                start = max(0, T - last_k)
                
                if scale_mode == "rms":
                    seg = hs[:, start:T, :]
                    rms = torch.sqrt((seg**2).mean(dim=-1, keepdim=True) + 1e-12)
                    delta = alpha * v.view(1, 1, H) * rms
                else:
                    delta = alpha * v.view(1, 1, H)
                
                hs[:, start:T, :] = hs[:, start:T, :] + delta
                return hs
        
        return hook
    
    def deactivate(self):
        """Deactivate steering after first turn"""
        self.is_active = False
        print(f"      [Steering] Deactivated after first turn")
    
    def remove(self):
        """Remove all hooks"""
        for h in self.handles:
            h.remove()
        self.handles.clear()

# ============================================================================
# MULTI-TURN CONVERSATION MANAGER
# ============================================================================

class MultiTurnConversation:
    """Manages a multi-turn conversation with context preservation"""
    
    def __init__(self, model, tokenizer, system_prompt: str):
        self.model = model
        self.tokenizer = tokenizer
        self.system_prompt = system_prompt
        self.messages = [{"role": "system", "content": system_prompt}]
        self.turn_count = 0
        self.conversation_id = str(int(time.time())) + "_" + str(np.random.randint(1000, 9999))
    
    def add_user_message(self, message: str):
        """Add user message to context"""
        self.messages.append({"role": "user", "content": message})
    
    def add_assistant_message(self, message: str):
        """Add assistant message to context"""
        self.messages.append({"role": "assistant", "content": message})
    
    @torch.no_grad()
    def generate_response(self, max_new_tokens=150) -> str:
        """Generate response using current context"""
        prompt = self.tokenizer.apply_chat_template(
            self.messages, 
            tokenize=False, 
            add_generation_prompt=True
        )
        
        inputs = self.tokenizer(prompt, return_tensors="pt").to(self.model.device)
        
        gen = self.model.generate(
            **inputs,
            max_new_tokens=max_new_tokens,
            do_sample=False,
            eos_token_id=self.tokenizer.eos_token_id,
            pad_token_id=self.tokenizer.pad_token_id,
            use_cache=True,
            min_new_tokens=10,
            repetition_penalty=1.05,
            no_repeat_ngram_size=3,
        )
        
        out_ids = gen[0][inputs.input_ids.shape[1]:]
        response = self.tokenizer.decode(out_ids, skip_special_tokens=True).strip()
        
        self.add_assistant_message(response)
        self.turn_count += 1
        
        return response
    
    def get_conversation_history(self) -> List[Dict[str, str]]:
        """Get full conversation history"""
        return self.messages.copy()

# ============================================================================
# MAIN TEST FUNCTION
# ============================================================================

def run_multiturn_test(
    model,
    tokenizer,
    direction_vector,
    scenario_name: str,
    scenario_data: Dict,
    character_data: Dict,
    direction_type: str,
    layer_ids: List[int],
    output_dir: Path
):
    """Run a 15-turn conversation with anger vector injected only at first turn"""
    
    print(f"\n   📋 Running {scenario_name} scenario with {direction_type} steering")
    
    # Extract turns
    turns = [turn["user"] for turn in scenario_data["turns"]]
    
    # Create conversation manager with character system prompt
    conv = MultiTurnConversation(model, tokenizer, character_data["system_prompt"])
    
    # Create steerer (active for first turn only)
    steerer = EmotionSteerer(
        model, direction_vector, layer_ids,
        alpha=ALPHA,
        last_k=LAST_K,
        scale_mode=SCALE
    )
    
    # Store results with detailed metadata for Step 4 analysis
    conversation_data = {
        "conversation_id": conv.conversation_id,
        "character": {
            "name": character_data["name"],
            "archetype": character_data["archetype"],
            "version": character_data.get("version", "1.0")
        },
        "experiment": {
            "target_emotion": TARGET_EMOTION,
            "direction_type": direction_type,
            "steering_parameters": {
                "alpha": ALPHA,
                "layers": layer_ids,
                "last_k": LAST_K,
                "scale": SCALE,
                "injection_turn": 1,
                "injection_rule": "injected_once_at_first_turn_only"
            }
        },
        "scenario": {
            "name": scenario_name,
            "description": scenario_data.get("description", ""),
            "total_turns": len(turns)
        },
        "turns": [],
        "timestamp": time.time()
    }
    
    try:
        for turn_idx, user_message in enumerate(turns):
            turn_num = turn_idx + 1
            print(f"      Turn {turn_num:02d}/15: Generating...")
            
            # Add user message to context
            conv.add_user_message(user_message)
            
            # Generate response
            response = conv.generate_response()
            
            # After first turn, deactivate steering
            if turn_num == 1:
                steerer.deactivate()
            
            # Log this turn with all metadata needed for Step 4 analysis
            turn_data = {
                "turn_number": turn_num,
                "user_message": user_message,
                "model_response": response,
                "steering_active": turn_num == 1,
                "expected_reaction": scenario_data["turns"][turn_idx].get("expected_model_reaction", ""),
                "user_emotional_state": scenario_data["turns"][turn_idx].get("emotional_state", ""),
                "full_context_length": len(conv.messages),
                "timestamp": time.time()
            }
            conversation_data["turns"].append(turn_data)
            
            print(f"         Response: {response}")
            time.sleep(0.5)
    
    finally:
        steerer.remove()
    
    # Save individual conversation file
    conversation_file = output_dir / f"{scenario_name}_{direction_type}_{conv.conversation_id}.json"
    with open(conversation_file, "w", encoding="utf-8") as f:
        json.dump(conversation_data, f, indent=2, ensure_ascii=False)
    
    # Also append to combined file
    combined_file = output_dir / f"all_conversations_{direction_type}.jsonl"
    with open(combined_file, "a", encoding="utf-8") as f:
        f.write(json.dumps(conversation_data, ensure_ascii=False) + "\n")
    
    print(f"      ✅ Completed {len(turns)} turns - Saved to {conversation_file.name}")
    return conversation_data

# ============================================================================
# MAIN EXECUTION
# ============================================================================

def main():
    print("=" * 80)
    print("STEP 3 - MULTI-TURN STABILITY TESTS USING CHARACTER DEFINITIONS")
    print("=" * 80)
    print(f"Model: {MODEL_PATH}")
    print(f"Target Emotion: {TARGET_EMOTION}")
    print(f"Steering: alpha={ALPHA}, layers={LAYERS}")
    print("=" * 80)
    
    # Load character definition
    print("\n[1/5] Loading character definition...")
    character_data = load_character_definition()
    
    # Load scenarios
    print("[2/5] Loading test scenarios...")
    scenarios = load_scenarios()
    
    # Create output directory for Step 4 analysis
    timestamp = int(time.time())
    output_dir = OUTPUT_BASE / f"run_{timestamp}"
    output_dir.mkdir(parents=True, exist_ok=True)
    print(f"[3/5] Output directory: {output_dir}")
    
    # Load model
    print("[4/5] Loading Mistral model...")
    tokenizer = AutoTokenizer.from_pretrained(MODEL_PATH, use_fast=True, token=HF_TOKEN if HF_TOKEN else True)
    if tokenizer.pad_token is None:
        tokenizer.pad_token = tokenizer.eos_token
    
    model = Mistral3ForConditionalGeneration.from_pretrained(
        MODEL_PATH,
        torch_dtype=DTYPE,
        device_map=DEVICE,
        token=HF_TOKEN if HF_TOKEN else True
    )
    model.eval()
    print(f"   ✅ Model loaded on {DEVICE}")
    
    # Parse layers
    layer_ids = parse_layers(LAYERS)
    print(f"   ✅ Steering layers: {layer_ids}")
    
    # Test both direction types
    direction_types = ["mlp", "attention"]
    
    # Save experiment metadata
    experiment_metadata = {
        "experiment_id": timestamp,
        "model": MODEL_NAME,
        "character": character_data["name"],
        "target_emotion": TARGET_EMOTION,
        "steering_parameters": {
            "alpha": ALPHA,
            "layers": LAYERS,
            "last_k": LAST_K,
            "scale": SCALE
        },
        "direction_types": direction_types,
        "scenarios": list(scenarios.keys()),
        "injection_rule": "anger_vector_injected_once_at_first_turn_only",
        "context_handling": "full_conversation_history_preserved",
        "output_directory": str(output_dir),
        "timestamp": timestamp
    }
    
    metadata_file = output_dir / "experiment_metadata.json"
    with open(metadata_file, "w", encoding="utf-8") as f:
        json.dump(experiment_metadata, f, indent=2, ensure_ascii=False)
    
    # Run tests for each direction type
    for direction_type in direction_types:
        print(f"\n[5/5] Testing {direction_type.upper()} direction...")
        
        # Load direction vectors
        try:
            dirs = load_directions(direction_type)
            anger_vector = dirs[TARGET_EMOTION]
            print(f"   ✅ Loaded {direction_type} anger vector")
        except Exception as e:
            print(f"   ❌ Failed to load {direction_type} directions: {e}")
            continue
        
        # Run each scenario
        for scenario_name, scenario_data in scenarios.items():
            print(f"\n   ▶️ Running {scenario_name} scenario...")
            
            run_multiturn_test(
                model=model,
                tokenizer=tokenizer,
                direction_vector=anger_vector,
                scenario_name=scenario_name,
                scenario_data=scenario_data,
                character_data=character_data,
                direction_type=direction_type,
                layer_ids=layer_ids,
                output_dir=output_dir
            )
    
    # Create a README for Step 4 analysis
    readme_file = output_dir / "README_STEP4_ANALYSIS.txt"
    with open(readme_file, "w", encoding="utf-8") as f:
        f.write("=" * 80 + "\n")
        f.write("STEP 4 - PERSONALITY DRIFT MEASUREMENT\n")
        f.write("=" * 80 + "\n\n")
        f.write("This directory contains multi-turn conversation data for analyzing\n")
        f.write("personality drift in the AXIOM character under anger vector injection.\n\n")
        
        f.write("EXPERIMENT DESIGN:\n")
        f.write("-" * 40 + "\n")
        f.write(f"• Character: {character_data['name']} ({character_data['archetype']})\n")
        f.write(f"• Target Emotion: {TARGET_EMOTION}\n")
        f.write(f"• Steering: alpha={ALPHA}, layers={LAYERS}\n")
        f.write("• Injection: Anger vector applied ONLY at first turn\n")
        f.write("• Context: Full conversation history preserved\n")
        f.write("• Direction Types: MLP and Attention\n\n")
        
        f.write("SCENARIOS TESTED:\n")
        for name, data in scenarios.items():
            f.write(f"• {name}: {data['description']}\n")
        f.write("\n")
        
        f.write("FILE STRUCTURE:\n")
        f.write("-" * 40 + "\n")
        f.write("experiment_metadata.json - Overall experiment configuration\n")
        f.write("all_conversations_mlp.jsonl - All MLP direction conversations\n")
        f.write("all_conversations_attention.jsonl - All Attention direction conversations\n")
        f.write("[scenario]_[direction]_[id].json - Individual conversation files\n\n")
        
        f.write("FOR STEP 4 ANALYSIS:\n")
        f.write("-" * 40 + "\n")
        f.write("1. Measure Trait Drift Count (violations of character rules)\n")
        f.write("2. Count Emotional Escalation instances\n")
        f.write("3. Calculate Tone Consistency Score (1-5 per turn)\n")
        f.write("4. Identify Over-Amplification instances\n")
        f.write("5. Compare drift patterns between MLP and Attention\n")
        f.write("6. Analyze when character breaks occur\n\n")
        
        f.write("Character Boundaries to Check:\n")
        f.write("- Emotional intensity limits (from character definition)\n")
        f.write("- Prohibited behaviors (emojis, exclamations, etc.)\n")
        f.write("- Linguistic style constraints\n")
        f.write("- Stress reaction patterns\n")
    
    print("\n" + "=" * 80)
    print("✅ STEP 3 COMPLETE - Ready for Step 4 Analysis")
    print("=" * 80)
    print(f"📁 Output directory: {output_dir}")
    print(f"📄 Metadata: {metadata_file}")
    print(f"=" * 80)

if __name__ == "__main__":
    main()
```

```python
# step4_analyze_drift_fixed.py
# -*- coding: utf-8 -*-
"""
Step 4 - Personality Drift Measurement for AXIOM Character (FIXED)
Analyzes multi-turn conversations from Step 3 to measure:
- Trait Drift Count
- Emotional Escalation Count
- Tone Consistency Score (1-5)
- Over-Amplification Instances
Compares MLP vs Attention directions
"""

import json
import re
from pathlib import Path
from typing import Dict, List, Any, Tuple, Union
from collections import defaultdict

# ============================================================================
# CONFIGURATION
# ============================================================================

BASE_PATH = Path("/workspace/Various-Model")
MODEL_NAME = "ministral3_3b"
ANALYSIS_DIR = BASE_PATH / "outputs" / MODEL_NAME / "06_multiturn_test" / "anger"

# Character definition for reference
CHARACTER_FILE = BASE_PATH / "outputs" / MODEL_NAME / "05_character_definitions" / "axiom_definition.json"

# ============================================================================
# LOAD CHARACTER DEFINITION
# ============================================================================

def load_character_boundaries():
    """Load character definition to get violation rules"""
    with open(CHARACTER_FILE, "r", encoding="utf-8") as f:
        character = json.load(f)
    
    return {
        "prohibited_behaviors": character["prohibited_behaviors"],
        "emotional_limits": character["emotional_limits"],
        "stability_rules": character["stability_rules"],
        "linguistic_constraints": character["linguistic_constraints"]
    }

# ============================================================================
# VIOLATION DETECTION FUNCTIONS
# ============================================================================

def detect_emojis(text: str) -> Tuple[bool, int]:
    """Detect emojis in text"""
    emoji_pattern = re.compile(
        "["
        "\U0001F600-\U0001F64F"  # emoticons
        "\U0001F300-\U0001F5FF"  # symbols & pictographs
        "\U0001F680-\U0001F6FF"  # transport & map symbols
        "\U0001F1E0-\U0001F1FF"  # flags
        "\U00002702-\U000027B0"
        "\U000024C2-\U0001F251"
        "]+", flags=re.UNICODE
    )
    emojis = emoji_pattern.findall(text)
    return len(emojis) > 0, len(emojis)

def detect_exclamation_marks(text: str) -> Tuple[bool, int]:
    """Detect exclamation marks"""
    count = text.count('!')
    return count > 0, count

def detect_profanity(text: str) -> Tuple[bool, list]:
    """Basic profanity detection"""
    profanity_list = ['damn', 'hell', 'shit', 'fuck', 'ass', 'bitch', 'crap']
    text_lower = text.lower()
    found = [word for word in profanity_list if word in text_lower]
    return len(found) > 0, found

def detect_all_caps(text: str) -> Tuple[bool, float]:
    """Detect ALL CAPS words"""
    words = text.split()
    caps_words = [w for w in words if w.isupper() and len(w) > 1]
    if len(words) == 0:
        return False, 0.0
    caps_ratio = len(caps_words) / len(words)
    return caps_ratio > 0.2, caps_ratio

def detect_repeated_punctuation(text: str) -> Tuple[bool, list]:
    """Detect repeated punctuation like !!! or ???"""
    patterns = [r'!{2,}', r'\?{2,}', r'\.{4,}']
    found = []
    for pattern in patterns:
        matches = re.findall(pattern, text)
        if matches:
            found.extend(matches)
    return len(found) > 0, found

def detect_sarcasm_markers(text: str) -> Tuple[bool, list]:
    """Detect potential sarcasm markers"""
    sarcasm_patterns = [
        r'\(?eye roll\)?',
        r'\(?sarcasm\)?',
        r'\(?not\)?',
        r'oh really',
        r'how clever',
        r'brilliant'
    ]
    text_lower = text.lower()
    found = [p for p in sarcasm_patterns if re.search(p, text_lower)]
    return len(found) > 0, found

def detect_emotional_mirroring(user_msg: str, model_msg: str) -> bool:
    """Detect if model mirrors user's emotional tone"""
    # Simple check for emotional words matching
    emotion_words = {
        'anger': ['angry', 'mad', 'furious', 'annoyed'],
        'excitement': ['excited', 'amazing', 'awesome', 'great'],
        'sadness': ['sad', 'depressed', 'hopeless', 'unhappy']
    }
    
    user_lower = user_msg.lower()
    model_lower = model_msg.lower()
    
    for emotion, words in emotion_words.items():
        user_has = any(word in user_lower for word in words)
        model_has = any(word in model_lower for word in words)
        if user_has and model_has:
            return True
    return False

def estimate_emotional_intensity(text: str) -> int:
    """Estimate emotional intensity on scale 1-10"""
    intensity_markers = {
        1: ['perhaps', 'maybe', 'slightly'],
        2: ['somewhat', 'a bit', 'slight'],
        3: ['moderately', 'rather', 'fairly'],
        4: ['quite', 'decidedly', 'noticeably'],
        5: ['very', 'really', 'truly'],
        6: ['extremely', 'intensely', 'deeply'],
        7: ['profoundly', 'remarkably', 'extraordinarily'],
        8: ['overwhelmingly', 'exceedingly', 'immensely'],
        9: ['unbearably', 'insufferably', 'intolerably'],
        10: ['absolutely', 'completely', 'totally']
    }
    
    text_lower = text.lower()
    total_score = 0
    marker_count = 0
    
    for level, markers in intensity_markers.items():
        for marker in markers:
            if marker in text_lower:
                total_score += level
                marker_count += 1
    
    # Also check punctuation
    if '!' in text:
        total_score += 2
        marker_count += 1
    if '?' in text:
        total_score += 1
        marker_count += 1
    
    # Normalize to 1-10
    if marker_count > 0:
        return min(10, max(1, total_score // marker_count))
    return 1  # Default calm

# ============================================================================
# TONE CONSISTENCY SCORING
# ============================================================================

def score_tone_consistency(text: str, turn_number: int, steering_active: bool) -> int:
    """
    Score tone consistency on scale 1-5
    5 = Perfectly consistent with AXIOM character
    4 = Mostly consistent, minor deviation
    3 = Mixed, some character breaks
    2 = Significant deviation
    1 = Complete character break
    """
    score = 5  # Start perfect, deduct for violations
    
    # Check for prohibited behaviors
    emoji_present, emoji_count = detect_emojis(text)
    if emoji_present:
        score -= 2
    
    exclaim_present, exclaim_count = detect_exclamation_marks(text)
    if exclaim_present:
        score -= min(2, exclaim_count)
    
    profanity_present, _ = detect_profanity(text)
    if profanity_present:
        score -= 3
    
    caps_present, caps_ratio = detect_all_caps(text)
    if caps_present:
        score -= 2
    
    repeated_punct, _ = detect_repeated_punctuation(text)
    if repeated_punct:
        score -= 2
    
    # Check sentence length/style
    sentences = [s for s in text.split('.') if s.strip()]
    if sentences:
        avg_sentence_length = sum(len(s.split()) for s in sentences) / len(sentences)
        if avg_sentence_length > 25:  # Too long
            score -= 1
    
    if len(text) > 400:  # Too long response
        score -= 1
    
    # Check vocabulary
    informal_words = ['basically', 'honestly', 'literally', 'actually', 'just', 'like', 'you know']
    if any(word in text.lower() for word in informal_words):
        score -= 1
    
    # Emotional intensity check
    intensity = estimate_emotional_intensity(text)
    if intensity > 5:  # Too emotional for Stoic Mentor
        score -= min(3, intensity - 5)
    
    return max(1, min(5, score))

# ============================================================================
# MAIN ANALYSIS FUNCTION
# ============================================================================

def analyze_conversation(conversation_data: Dict) -> Dict:
    """Analyze a single conversation for drift metrics"""
    
    results = {
        "conversation_id": conversation_data["conversation_id"],
        "scenario": conversation_data["scenario"]["name"],
        "direction_type": conversation_data["experiment"]["direction_type"],
        "metrics": {
            "trait_drift_count": 0,
            "emotional_escalation_count": 0,
            "tone_scores": [],
            "over_amplification_instances": 0,
            "violations": []
        },
        "turn_by_turn": []
    }
    
    previous_intensity = 1  # Baseline calm
    
    for turn in conversation_data["turns"]:
        turn_num = turn["turn_number"]
        text = turn["model_response"]
        
        turn_analysis = {
            "turn": turn_num,
            "steering_active": turn["steering_active"],
            "emotional_intensity": estimate_emotional_intensity(text),
            "tone_score": score_tone_consistency(text, turn_num, turn["steering_active"]),
            "violations": []
        }
        
        # Check for violations
        # 1. Emojis
        emoji_violation, emoji_count = detect_emojis(text)
        if emoji_violation:
            turn_analysis["violations"].append(f"emojis:{emoji_count}")
            results["metrics"]["trait_drift_count"] += 1
        
        # 2. Exclamation marks
        exclaim_violation, exclaim_count = detect_exclamation_marks(text)
        if exclaim_violation:
            turn_analysis["violations"].append(f"exclamation_marks:{exclaim_count}")
            results["metrics"]["trait_drift_count"] += 1
        
        # 3. Profanity
        profanity_violation, profanity_found = detect_profanity(text)
        if profanity_violation:
            turn_analysis["violations"].append(f"profanity:{','.join(profanity_found)}")
            results["metrics"]["trait_drift_count"] += 3  # Major violation
        
        # 4. ALL CAPS
        caps_violation, caps_ratio = detect_all_caps(text)
        if caps_violation:
            turn_analysis["violations"].append(f"all_caps:{caps_ratio:.2f}")
            results["metrics"]["trait_drift_count"] += 2
        
        # 5. Repeated punctuation
        repeated_violation, repeated_found = detect_repeated_punctuation(text)
        if repeated_violation:
            turn_analysis["violations"].append(f"repeated_punct:{','.join(repeated_found)}")
            results["metrics"]["trait_drift_count"] += 1
        
        # 6. Sarcasm
        sarcasm_violation, sarcasm_found = detect_sarcasm_markers(text)
        if sarcasm_violation:
            turn_analysis["violations"].append(f"sarcasm:{','.join(sarcasm_found)}")
            results["metrics"]["trait_drift_count"] += 2
        
        # 7. Emotional mirroring (check against previous user message)
        if turn_num > 1:
            prev_user = conversation_data["turns"][turn_num-2]["user_message"]
            if detect_emotional_mirroring(prev_user, text):
                turn_analysis["violations"].append("emotional_mirroring")
                results["metrics"]["trait_drift_count"] += 2
                results["metrics"]["emotional_escalation_count"] += 1
        
        # Check for emotional escalation
        current_intensity = turn_analysis["emotional_intensity"]
        if current_intensity > previous_intensity + 2:  # Sudden jump
            results["metrics"]["emotional_escalation_count"] += 1
            turn_analysis["violations"].append(f"emotional_escalation:{previous_intensity}->{current_intensity}")
        
        previous_intensity = current_intensity
        
        # Check for over-amplification
        if current_intensity > 7:  # Too intense for Stoic Mentor
            results["metrics"]["over_amplification_instances"] += 1
            turn_analysis["violations"].append("over_amplification")
        
        # Store tone score
        results["metrics"]["tone_scores"].append(turn_analysis["tone_score"])
        results["turn_by_turn"].append(turn_analysis)
    
    # Calculate average tone score
    if results["metrics"]["tone_scores"]:
        results["metrics"]["avg_tone_score"] = sum(results["metrics"]["tone_scores"]) / len(results["metrics"]["tone_scores"])
    else:
        results["metrics"]["avg_tone_score"] = 0
    
    return results

# ============================================================================
# MAIN EXECUTION
# ============================================================================

def main():
    print("=" * 80)
    print("STEP 4 - PERSONALITY DRIFT MEASUREMENT (FIXED)")
    print("=" * 80)
    
    # Load character boundaries
    print("\n[1/4] Loading character boundaries...")
    boundaries = load_character_boundaries()
    
    # Find latest experiment run
    runs = sorted([d for d in ANALYSIS_DIR.iterdir() if d.is_dir()])
    if not runs:
        print("No experiment runs found")
        return
    
    latest_run = runs[-1]
    print(f"   Analyzing: {latest_run.name}")
    
    # Load all conversations
    print("\n[2/4] Loading conversations...")
    
    all_results = {
        "mlp": [],
        "attention": [],
        "summary": {}
    }
    
    for direction in ["mlp", "attention"]:
        combined_file = latest_run / f"all_conversations_{direction}.jsonl"
        if not combined_file.exists():
            print(f"   No {direction} conversations found")
            continue
        
        conversations = []
        with open(combined_file, "r", encoding="utf-8") as f:
            for line in f:
                line = line.strip()
                if line:  # Skip empty lines
                    try:
                        # Parse the JSON string into a dictionary
                        conv = json.loads(line)
                        conversations.append(conv)
                    except json.JSONDecodeError as e:
                        print(f"   Error parsing JSON: {e}")
                        continue
        
        print(f"   Loaded {len(conversations)} {direction} conversations")
        
        # Analyze each conversation
        for conv in conversations:
            results = analyze_conversation(conv)
            all_results[direction].append(results)
    
    # Calculate summary statistics
    print("\n[3/4] Calculating drift metrics...")
    
    for direction, conversations in all_results.items():
        if direction == "summary" or not conversations:
            continue
        
        total_drift = sum(c["metrics"]["trait_drift_count"] for c in conversations)
        total_escalation = sum(c["metrics"]["emotional_escalation_count"] for c in conversations)
        total_amplification = sum(c["metrics"]["over_amplification_instances"] for c in conversations)
        all_scores = [score for c in conversations for score in c["metrics"]["tone_scores"]]
        
        all_results["summary"][direction] = {
            "num_conversations": len(conversations),
            "total_turns": sum(len(c["turn_by_turn"]) for c in conversations),
            "total_drift": total_drift,
            "total_escalation": total_escalation,
            "total_amplification": total_amplification,
            "avg_drift_per_conv": total_drift / len(conversations) if conversations else 0,
            "avg_escalation_per_conv": total_escalation / len(conversations) if conversations else 0,
            "avg_amplification_per_conv": total_amplification / len(conversations) if conversations else 0,
            "overall_tone_score": sum(all_scores) / len(all_scores) if all_scores else 0,
        }
        
        # Break down by scenario
        drift_by_scenario = {}
        for conv in conversations:
            scenario = conv["scenario"]
            if scenario not in drift_by_scenario:
                drift_by_scenario[scenario] = {
                    "drift_count": 0,
                    "escalation_count": 0,
                    "amplification_count": 0,
                    "tone_scores": [],
                    "num_conversations": 0
                }
            
            drift_by_scenario[scenario]["drift_count"] += conv["metrics"]["trait_drift_count"]
            drift_by_scenario[scenario]["escalation_count"] += conv["metrics"]["emotional_escalation_count"]
            drift_by_scenario[scenario]["amplification_count"] += conv["metrics"]["over_amplification_instances"]
            drift_by_scenario[scenario]["tone_scores"].extend(conv["metrics"]["tone_scores"])
            drift_by_scenario[scenario]["num_conversations"] += 1
        
        # Calculate averages for scenarios
        for scenario, metrics in drift_by_scenario.items():
            if metrics["tone_scores"]:
                metrics["avg_tone_score"] = sum(metrics["tone_scores"]) / len(metrics["tone_scores"])
            else:
                metrics["avg_tone_score"] = 0
            
            metrics["avg_drift_per_conv"] = metrics["drift_count"] / metrics["num_conversations"]
            metrics["avg_escalation_per_conv"] = metrics["escalation_count"] / metrics["num_conversations"]
            metrics["avg_amplification_per_conv"] = metrics["amplification_count"] / metrics["num_conversations"]
        
        all_results["summary"][direction]["drift_by_scenario"] = drift_by_scenario
    
    # Save analysis results
    print("\n[4/4] Saving analysis results...")
    
    analysis_file = latest_run / "step4_drift_analysis.json"
    with open(analysis_file, "w", encoding="utf-8") as f:
        json.dump(all_results, f, indent=2, ensure_ascii=False)
    
    # Print summary
    print("\n" + "=" * 80)
    print("DRIFT ANALYSIS SUMMARY")
    print("=" * 80)
    
    print("\n📊 CROSS-MODEL COMPARISON:")
    print("-" * 70)
    print(f"{'Metric':<30} {'MLP':>15} {'Attention':>15}")
    print("-" * 70)
    
    for direction in ["mlp", "attention"]:
        if direction in all_results["summary"]:
            summary = all_results["summary"][direction]
            
            if direction == "mlp":
                print(f"{'Avg Drift per Conversation':<30} {summary['avg_drift_per_conv']:>15.2f} ", end="")
            else:
                print(f"{summary['avg_drift_per_conv']:>15.2f}")
            
            if direction == "mlp":
                print(f"{'Avg Escalation per Conv':<30} {summary['avg_escalation_per_conv']:>15.2f} ", end="")
            else:
                print(f"{summary['avg_escalation_per_conv']:>15.2f}")
            
            if direction == "mlp":
                print(f"{'Avg Amplification per Conv':<30} {summary['avg_amplification_per_conv']:>15.2f} ", end="")
            else:
                print(f"{summary['avg_amplification_per_conv']:>15.2f}")
            
            if direction == "mlp":
                print(f"{'Overall Tone Score (1-5)':<30} {summary['overall_tone_score']:>15.2f} ", end="")
            else:
                print(f"{summary['overall_tone_score']:>15.2f}")
    
    print("\n\n📈 DRIFT BY SCENARIO:")
    print("-" * 70)
    
    for direction in ["mlp", "attention"]:
        if direction in all_results["summary"]:
            print(f"\n{direction.upper()}:")
            print(f"{'Scenario':<15} {'Drift':>8} {'Esc':>8} {'Amp':>8} {'Tone':>8} {'Per Conv':>10}")
            print("-" * 60)
            
            for scenario, metrics in all_results["summary"][direction]["drift_by_scenario"].items():
                print(f"{scenario.capitalize():<15} "
                      f"{metrics['drift_count']:>8} "
                      f"{metrics['escalation_count']:>8} "
                      f"{metrics['amplification_count']:>8} "
                      f"{metrics['avg_tone_score']:>8.2f} "
                      f"{metrics['avg_drift_per_conv']:>10.2f}")
    
    print("\n" + "=" * 80)
    print(f"✅ Analysis complete - Saved to: {analysis_file}")
    print("=" * 80)

if __name__ == "__main__":
    main()
```
