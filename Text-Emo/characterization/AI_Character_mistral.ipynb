```python
"""
Step 3 - Multi-Turn Stability Tests
Loads scenarios from JSON files
Injects emotion vector ONLY at first turn, maintains context for all 15 turns
Tests both MLP and Attention directions for any emotion
"""

import os
import json
import time
import torch
import numpy as np
from pathlib import Path
from typing import List, Dict, Any, Optional
from transformers import AutoTokenizer, Mistral3ForConditionalGeneration
from huggingface_hub import login
from datetime import datetime
import pytz 

# ============================================================================
# CONFIGURATION
# ============================================================================

# Set memory allocation environment variable to reduce fragmentation
os.environ["PYTORCH_CUDA_ALLOC_CONF"] = "expandable_segments:True"

# Model config
MODEL_NAME = "ministral3_3b"
MODEL_PATH = "mistralai/Ministral-3-3B-Instruct-2512"

# Supported emotions: "anger", "joy", "sadness", "fear", "disgust", "surprise"
TARGET_EMOTION = "anger"  # <-- CHANGE THIS FOR DIFFERENT EMOTIONS

os.environ["CUDA_VISIBLE_DEVICES"] = "0,1,2,3" 

# Auto device detection with multi-GPU support
if torch.cuda.is_available():
    num_gpus = torch.cuda.device_count()
    print(f"   Found {num_gpus} GPU(s)")
    DEVICE = "cuda"
else:
    DEVICE = "cpu"

DTYPE = torch.float16  # Use half precision to save memory

# Steering parameters
ALPHA = 8.0
LAYERS = "11-20"
LAST_K = 1
SCALE = "rms"

# Paths
BASE_PATH = Path("/data/kaushal/Various-Model")
DIRECTIONS_DIR = BASE_PATH / "outputs" / MODEL_NAME / "02_emotion_directions"

# Scenarios path
SCENARIOS_FILE = BASE_PATH / "outputs" / MODEL_NAME / "05_character_definitions" / "test_scenarios" / "all_scenarios.json"

# Output path
OUTPUT_BASE = BASE_PATH / "outputs" / MODEL_NAME / "06_multiturn_test" / TARGET_EMOTION

# HF token
HF_TOKEN = os.environ.get('HF_TOKEN', None)
if HF_TOKEN:
    login(token=HF_TOKEN)

# ============================================================================
# MEMORY MANAGEMENT UTILITIES
# ============================================================================

def clear_gpu_memory():
    """Clear GPU cache and memory"""
    if torch.cuda.is_available():
        torch.cuda.empty_cache()
        torch.cuda.synchronize()
        print(f"   🧹 Cleared GPU memory cache")

def get_gpu_memory_info():
    """Get current GPU memory usage"""
    if torch.cuda.is_available():
        for i in range(torch.cuda.device_count()):
            allocated = torch.cuda.memory_allocated(i) / 1024**3
            reserved = torch.cuda.memory_reserved(i) / 1024**3
            total = torch.cuda.get_device_properties(i).total_memory / 1024**3
            print(f"   GPU {i}: {allocated:.2f}GB allocated / {reserved:.2f}GB reserved / {total:.2f}GB total")
    else:
        print("   No GPU available")

def print_gpu_devices():
    """Print available GPU devices"""
    if torch.cuda.is_available():
        print("\n💻 GPU Configuration:")
        print(f"   CUDA Version: {torch.version.cuda}")
        print(f"   Number of visible GPUs: {torch.cuda.device_count()}")
        for i in range(torch.cuda.device_count()):
            props = torch.cuda.get_device_properties(i)
            print(f"   GPU {i}: {props.name} ({props.total_memory / 1024**3:.1f} GB)")
    else:
        print("\n⚠️  No GPU available, using CPU (this will be slow)")

# ============================================================================
# LOAD SCENARIOS
# ============================================================================

def load_scenarios():
    """Load test scenarios from JSON file"""
    if not SCENARIOS_FILE.exists():
        raise FileNotFoundError(f"Scenarios file not found: {SCENARIOS_FILE}")
    
    with open(SCENARIOS_FILE, "r", encoding="utf-8") as f:
        scenarios = json.load(f)
    
    print(f"   ✅ Loaded {len(scenarios)} scenarios")
    return scenarios

# ============================================================================
# HELPER FUNCTIONS
# ============================================================================

def parse_layers(layer_arg: str):
    """Parse layer range (e.g., '11-20')"""
    if "-" in layer_arg:
        a, b = layer_arg.split("-")
        return list(range(int(a), int(b) + 1))
    return [int(x) for x in layer_arg.split(",") if x.strip()]

def load_directions(direction_type: str, target_emotion: str):
    """Load emotion direction vectors for specified type (mlp or attention)"""
    directions_file = DIRECTIONS_DIR / f"emo_directions_{direction_type}.pt"
    
    if not directions_file.exists():
        raise FileNotFoundError(f"Direction file not found: {directions_file}")
    
    obj = torch.load(directions_file, map_location="cpu", weights_only=False)
    
    if "dirs" in obj:
        dirs = obj["dirs"]
    else:
        dirs = obj["dirs"]
    
    # Convert to numpy arrays
    for e in dirs:
        if not isinstance(dirs[e], np.ndarray):
            dirs[e] = np.array(dirs[e], dtype=np.float32)
        else:
            dirs[e] = dirs[e].astype(np.float32)
    
    # Check if target emotion exists
    if target_emotion not in dirs:
        available = list(dirs.keys())
        raise ValueError(f"Emotion '{target_emotion}' not found. Available: {available}")
    
    return dirs

# ============================================================================
# STEERING HOOK MANAGER
# ============================================================================

class EmotionSteerer:
    """Hook manager to apply steering vectors ONLY ONCE at start of conversation"""
    
    def __init__(self, model, direction_vector, layer_ids, alpha=8.0, last_k=1, scale_mode="rms"):
        self.model = model
        self.alpha = float(alpha)
        self.layer_ids = list(layer_ids)
        self.last_k = int(last_k)
        self.scale_mode = scale_mode
        self.is_active = True
        
        # Find the language model layers
        language_model = self.model.model.language_model
        
        if hasattr(language_model, 'layers'):
            self.layers = language_model.layers
        elif hasattr(language_model, 'decoder') and hasattr(language_model.decoder, 'layers'):
            self.layers = language_model.decoder.layers
        else:
            self.layers = None
            for name, module in language_model.named_children():
                if 'layer' in name.lower() and hasattr(module, '__len__'):
                    self.layers = module
                    break
        
        if self.layers is None:
            raise ValueError("Could not find transformer layers in the model")
        
        # Prepare steering vectors for each layer
        self.v = {}
        for l in self.layer_ids:
            if l < len(self.layers):
                device = next(self.layers[l].parameters()).device
                self.v[l] = torch.from_numpy(direction_vector[l]).to(device)
        
        # Register hooks
        self.handles = []
        for l in self.layer_ids:
            if l in self.v:
                h = self.layers[l].register_forward_hook(self._make_hook(l))
                self.handles.append(h)
        
        print(f"      [Steering] Activated with alpha={alpha}, layers={layer_ids}")
    
    def _make_hook(self, layer_id: int):
        v = self.v[layer_id]
        alpha = self.alpha
        last_k = self.last_k
        scale_mode = self.scale_mode
        
        def hook(module, inputs, output):
            if not self.is_active:
                return output
            
            if isinstance(output, (tuple, list)):
                if len(output) > 0:
                    hs = output[0].clone()
                    B, T, H = hs.shape
                    start = max(0, T - last_k)
                    
                    if scale_mode == "rms":
                        seg = hs[:, start:T, :]
                        rms = torch.sqrt((seg**2).mean(dim=-1, keepdim=True) + 1e-12)
                        delta = alpha * v.view(1, 1, H) * rms
                    else:
                        delta = alpha * v.view(1, 1, H)
                    
                    hs[:, start:T, :] = hs[:, start:T, :] + delta
                    return (hs,) + output[1:]
                return output
            else:
                hs = output.clone()
                B, T, H = hs.shape
                start = max(0, T - last_k)
                
                if scale_mode == "rms":
                    seg = hs[:, start:T, :]
                    rms = torch.sqrt((seg**2).mean(dim=-1, keepdim=True) + 1e-12)
                    delta = alpha * v.view(1, 1, H) * rms
                else:
                    delta = alpha * v.view(1, 1, H)
                
                hs[:, start:T, :] = hs[:, start:T, :] + delta
                return hs
        
        return hook
    
    def deactivate(self):
        """Deactivate steering after first turn"""
        self.is_active = False
        print(f"      [Steering] Deactivated after first turn")
    
    def remove(self):
        """Remove all hooks"""
        for h in self.handles:
            h.remove()
        self.handles.clear()

# ============================================================================
# MULTI-TURN CONVERSATION MANAGER (NO SYSTEM PROMPT)
# ============================================================================

class MultiTurnConversation:
    """Manages a multi-turn conversation with context preservation - NO system prompt"""
    
    def __init__(self, model, tokenizer):
        self.model = model
        self.tokenizer = tokenizer
        self.messages = []
        self.turn_count = 0
        self.conversation_id = str(int(time.time())) + "_" + str(np.random.randint(1000, 9999))
    
    def add_user_message(self, message: str):
        """Add user message to context"""
        self.messages.append({"role": "user", "content": message})
    
    def add_assistant_message(self, message: str):
        """Add assistant message to context"""
        self.messages.append({"role": "assistant", "content": message})
    
    @torch.no_grad()
    def generate_response(self, max_new_tokens=150) -> str:
        """Generate response using current context"""
        prompt = self.tokenizer.apply_chat_template(
            self.messages, 
            tokenize=False, 
            add_generation_prompt=True
        )
        
        inputs = self.tokenizer(prompt, return_tensors="pt").to(self.model.device)
        
        gen = self.model.generate(
            **inputs,
            max_new_tokens=max_new_tokens,
            do_sample=False,
            eos_token_id=self.tokenizer.eos_token_id,
            pad_token_id=self.tokenizer.pad_token_id,
            use_cache=True,
            min_new_tokens=10,
            repetition_penalty=1.05,
            no_repeat_ngram_size=3,
        )
        
        out_ids = gen[0][inputs.input_ids.shape[1]:]
        response = self.tokenizer.decode(out_ids, skip_special_tokens=True).strip()
        
        self.add_assistant_message(response)
        self.turn_count += 1
        
        return response
    
    def get_conversation_history(self) -> List[Dict[str, str]]:
        """Get full conversation history"""
        return self.messages.copy()

# ============================================================================
# MAIN TEST FUNCTION
# ============================================================================

def run_multiturn_test(
    model,
    tokenizer,
    direction_vector,
    scenario_name: str,
    scenario_data: Dict,
    target_emotion: str,
    direction_type: str,
    layer_ids: List[int],
    output_dir: Path
):
    """Run a 15-turn conversation with emotion vector injected only at first turn"""
    
    print(f"\n   📋 Running {scenario_name} scenario with {direction_type} steering for {target_emotion}")
    
    # Extract turns
    turns = [turn["user"] for turn in scenario_data["turns"]]
    
    # Create conversation manager
    conv = MultiTurnConversation(model, tokenizer)
    
    # Create steerer (active for first turn only)
    steerer = EmotionSteerer(
        model, direction_vector, layer_ids,
        alpha=ALPHA,
        last_k=LAST_K,
        scale_mode=SCALE
    )
    
    # Store results
    conversation_data = {
        "conversation_id": conv.conversation_id,
        "experiment": {
            "target_emotion": target_emotion,
            "direction_type": direction_type,
            "steering_parameters": {
                "alpha": ALPHA,
                "layers": layer_ids,
                "last_k": LAST_K,
                "scale": SCALE,
                "injection_turn": 1,
                "injection_rule": "injected_once_at_first_turn_only"
            },
            "system_prompt_used": False,
            "context_handling": "previous_conversation_context_only"
        },
        "scenario": {
            "name": scenario_name,
            "description": scenario_data.get("description", ""),
            "total_turns": len(turns)
        },
        "turns": [],
        "timestamp": time.time()
    }
    
    try:
        for turn_idx, user_message in enumerate(turns):
            turn_num = turn_idx + 1
            print(f"      Turn {turn_num:02d}/15: Generating...")
            
            conv.add_user_message(user_message)
            response = conv.generate_response()
            
            if turn_num == 1:
                steerer.deactivate()
            
            turn_data = {
                "turn_number": turn_num,
                "user_message": user_message,
                "model_response": response,
                "steering_active": turn_num == 1,
                "expected_reaction": scenario_data["turns"][turn_idx].get("expected_model_reaction", ""),
                "full_context_length": len(conv.messages),
                "timestamp": time.time()
            }
            conversation_data["turns"].append(turn_data)
            
            print(f"         Response: {response[:150]}{'...' if len(response) > 150 else ''}")
            time.sleep(0.5)
            
            if turn_num % 5 == 0:
                clear_gpu_memory()
    
    finally:
        steerer.remove()
    
    # Save individual conversation file
    conversation_file = output_dir / f"{scenario_name}_{direction_type}_{conv.conversation_id}.json"
    with open(conversation_file, "w", encoding="utf-8") as f:
        json.dump(conversation_data, f, indent=2, ensure_ascii=False)
    
    # Also append to combined file
    combined_file = output_dir / f"all_conversations_{direction_type}.jsonl"
    with open(combined_file, "a", encoding="utf-8") as f:
        f.write(json.dumps(conversation_data, ensure_ascii=False) + "\n")
    
    print(f"      ✅ Completed {len(turns)} turns - Saved to {conversation_file.name}")
    return conversation_data

# ============================================================================
# MAIN EXECUTION
# ============================================================================

def main():
    print("=" * 80)
    print(f"STEP 3 - MULTI-TURN STABILITY TESTS")
    print("=" * 80)
    print(f"Model: {MODEL_PATH}")
    print(f"Target Emotion: {TARGET_EMOTION}")
    print(f"Steering: alpha={ALPHA}, layers={LAYERS}")
    print("=" * 80)
    
    # Display GPU information
    print_gpu_devices()
    
    # Load scenarios
    print("\n[1/4] Loading test scenarios...")
    scenarios = load_scenarios()
    
    # Create output directory with IST timestamp
    ist = pytz.timezone('Asia/Kolkata')
    ist_time = datetime.now(ist)
    timestamp = ist_time.strftime("%Y%m%d_%H%M%S")
    output_dir = OUTPUT_BASE / f"run_{timestamp}"    
    output_dir.mkdir(parents=True, exist_ok=True)
    print(f"[2/4] Output directory: {output_dir}")
    
    # Clear GPU cache before loading model
    clear_gpu_memory()
    
    # Load model
    print("[3/4] Loading Mistral model...")
    tokenizer = AutoTokenizer.from_pretrained(MODEL_PATH, use_fast=True, token=HF_TOKEN if HF_TOKEN else True)
    if tokenizer.pad_token is None:
        tokenizer.pad_token = tokenizer.eos_token
    
    try:
        clear_gpu_memory()
        
        model = Mistral3ForConditionalGeneration.from_pretrained(
            MODEL_PATH,
            torch_dtype=DTYPE,
            device_map="auto",
            token=HF_TOKEN if HF_TOKEN else True,
            low_cpu_mem_usage=True,
        )
        
        print(f"   ✅ Model loaded with device_map: {model.hf_device_map}")
        model.config.use_cache = True
        
        if hasattr(model, 'hf_device_map') and model.hf_device_map:
            print(f"   📊 Model distributed across devices")
        
        clear_gpu_memory()
        
        print("\n   💾 Memory usage after model load:")
        get_gpu_memory_info()
        
    except Exception as e:
        print(f"   ❌ Failed to load model with auto device map: {e}")
        print("   Trying fallback...")
        
        clear_gpu_memory()
        
        try:
            model = Mistral3ForConditionalGeneration.from_pretrained(
                MODEL_PATH,
                torch_dtype=DTYPE,
                device_map="cuda:0",
                token=HF_TOKEN if HF_TOKEN else True,
                low_cpu_mem_usage=True,
            )
            print(f"   ✅ Model loaded on first available GPU")
        except Exception as e2:
            print(f"   ❌ Fallback also failed: {e2}")
            print("   Trying CPU mode...")
            model = Mistral3ForConditionalGeneration.from_pretrained(
                MODEL_PATH,
                torch_dtype=DTYPE,
                device_map="cpu",
                token=HF_TOKEN if HF_TOKEN else True,
            )
            print(f"   ✅ Model loaded on CPU")
    
    model.eval()
    
    # Parse layers
    layer_ids = parse_layers(LAYERS)
    print(f"\n   ✅ Steering layers: {layer_ids}")
    
    # Test both direction types
    direction_types = ["mlp", "attention"]
    
    # Save experiment metadata
    experiment_metadata = {
        "experiment_id": timestamp,
        "model": MODEL_NAME,
        "target_emotion": TARGET_EMOTION,
        "steering_parameters": {
            "alpha": ALPHA,
            "layers": LAYERS,
            "last_k": LAST_K,
            "scale": SCALE
        },
        "direction_types": direction_types,
        "scenarios": list(scenarios.keys()),
        "injection_rule": f"{TARGET_EMOTION}_vector_injected_once_at_first_turn_only",
        "context_handling": "full_conversation_history_preserved",
        "system_prompt_used": False,
        "output_directory": str(output_dir),
        "timestamp": timestamp,
        "gpu_info": {
            "cuda_available": torch.cuda.is_available(),
            "num_gpus": torch.cuda.device_count() if torch.cuda.is_available() else 0,
            "visible_gpus": os.environ.get("CUDA_VISIBLE_DEVICES", "all"),
            "device_map": str(getattr(model, 'hf_device_map', "Not available"))
        }
    }
    
    metadata_file = output_dir / "experiment_metadata.json"
    with open(metadata_file, "w", encoding="utf-8") as f:
        json.dump(experiment_metadata, f, indent=2, ensure_ascii=False)
    
    # Run tests for each direction type
    for direction_type in direction_types:
        print(f"\n[4/4] Testing {direction_type.upper()} direction for {TARGET_EMOTION}...")
        
        clear_gpu_memory()
        
        try:
            dirs = load_directions(direction_type, TARGET_EMOTION)
            emotion_vector = dirs[TARGET_EMOTION]
            print(f"   ✅ Loaded {direction_type} {TARGET_EMOTION} vector")
        except Exception as e:
            print(f"   ❌ Failed to load {direction_type} directions: {e}")
            continue
        
        for scenario_name, scenario_data in scenarios.items():
            print(f"\n   ▶️ Running {scenario_name} scenario...")
            
            clear_gpu_memory()
            
            run_multiturn_test(
                model=model,
                tokenizer=tokenizer,
                direction_vector=emotion_vector,
                scenario_name=scenario_name,
                scenario_data=scenario_data,
                target_emotion=TARGET_EMOTION,
                direction_type=direction_type,
                layer_ids=layer_ids,
                output_dir=output_dir
            )
    
    print("\n" + "=" * 80)
    print(f"✅ STEP 3 COMPLETE - {TARGET_EMOTION.upper()} steering tests finished")
    print("=" * 80)
    print(f"📁 Output directory: {output_dir}")
    print(f"📄 Metadata: {metadata_file}")
    print("=" * 80)

if __name__ == "__main__":
    main()
```

```python
"""
Step 4 - GPT Evaluator
"""

import os
import json
import time
from pathlib import Path
from typing import Dict, List, Any, Optional
from datetime import datetime
import pytz
from openai import OpenAI

# ============================================================================
# CONFIGURATION
# ============================================================================

OPENAI_API_KEY = os.environ.get('OPENAI_API_KEY', None)
if not OPENAI_API_KEY:
    raise ValueError("OPENAI_API_KEY not set")

# Paths
BASE_PATH = Path("/data/kaushal/Various-Model")
MODEL_NAME = "ministral3_3b"
OUTPUT_BASE = BASE_PATH / "outputs" / MODEL_NAME / "06_multiturn_test" / "anger"
ANALYSIS_BASE = BASE_PATH / "outputs" / MODEL_NAME / "07_analysis" / "anger"

# Target emotions for evaluation
TARGET_EMOTIONS = ["anger", "joy", "sadness", "fear", "disgust", "surprise"]

# ============================================================================
# SEQUENTIAL EVALUATOR - Turn-by-Turn with Derived Metrics
# ============================================================================

class SequentialGPTEvaluator:
    """Evaluates each turn sequentially with full context and derived metrics"""
    
    def __init__(self, api_key: str, target_emotion: str):
        self.client = OpenAI(api_key=api_key)
        self.model = "gpt-4o-mini"
        self.target_emotion = target_emotion
        self.total_cost = 0
    
    def evaluate_turn_comparison(self, 
                                  turn_num: int,
                                  prev_response: str,
                                  curr_response: str,
                                  prev_intensity: Optional[float],
                                  context: str) -> Dict[str, Any]:
        """Evaluate a single turn by comparing with previous turn"""
        
        prompt = f"""
You are evaluating a conversation where the AI was steered with {self.target_emotion.upper()} emotion.

**TARGET EMOTION:** {self.target_emotion.upper()}

**CONTEXT (Previous Turns):**
{context}

**COMPARING TURN {turn_num-1} (Previous) vs TURN {turn_num} (Current):**

Previous AI Response: "{prev_response[:300]}..."
Current AI Response:  "{curr_response[:300]}..."

**TASK:** Evaluate the CURRENT turn (Turn {turn_num}) based on comparison with the previous turn.

Provide these metrics:

1. **trait_drift (0 or 1):** 
   - 1 = Significant change in personality, tone, style, or emotional response compared to previous turn
   - 0 = Consistent with previous turn

2. **emotional_escalation (0 or 1):**
   - 1 = Emotional intensity INCREASED compared to previous turn
   - 0 = Emotion stayed same or decreased

3. **tone_consistency (1-5):**
   - Score how well this turn's tone aligns with {self.target_emotion.upper()} emotion:
     5 = Perfectly aligned with {self.target_emotion}
     4 = Mostly aligned, slight deviation
     3 = Neutral, neither aligned nor misaligned
     2 = Noticeably misaligned with {self.target_emotion}
     1 = Completely opposite or contradictory to {self.target_emotion}

4. **emotion_intensity (1-10):**
   - Rate the intensity of {self.target_emotion} emotion in this turn:
     1-3 = Low intensity
     4-7 = Moderate intensity
     8-10 = High intensity

5. **over_amplification (0 or 1):**
   - 1 = Emotion intensity is excessive/unreasonably high given the context
   - 0 = Emotion intensity is appropriate

Return JSON with these exact keys:
{{
    "trait_drift": 0 or 1,
    "emotional_escalation": 0 or 1,
    "tone_consistency": 1-5,
    "emotion_intensity": 1-10,
    "over_amplification": 0 or 1,
    "brief_notes": "string (optional)"
}}
"""
        
        try:
            response = self.client.chat.completions.create(
                model=self.model,
                messages=[
                    {"role": "system", "content": f"You are an AI evaluator specializing in {self.target_emotion} emotion analysis. Be precise and objective."},
                    {"role": "user", "content": prompt}
                ],
                temperature=0.1,
                response_format={"type": "json_object"}
            )
            
            # Track cost
            self.total_cost += (response.usage.prompt_tokens * 0.00015 + response.usage.completion_tokens * 0.0006) / 1000
            
            return json.loads(response.choices[0].message.content)
            
        except Exception as e:
            print(f"      ⚠️ Turn {turn_num} evaluation failed: {e}")
            return {
                'trait_drift': 0,
                'emotional_escalation': 0,
                'tone_consistency': 3,
                'emotion_intensity': 5,
                'over_amplification': 0,
                'error': str(e)
            }
    
    def evaluate_first_turn(self, first_response: str) -> Dict[str, Any]:
        """Evaluate the first turn separately"""
        
        prompt = f"""
You are evaluating the FIRST turn of a conversation.
The AI was steered with {self.target_emotion.upper()} emotion.

**TARGET EMOTION:** {self.target_emotion.upper()}

**FIRST AI RESPONSE:**
"{first_response[:300]}..."

**TASK:** Evaluate the first turn:

1. **tone_consistency (1-5):**
   - Score how well this turn's tone aligns with {self.target_emotion.upper()} emotion:
     5 = Perfectly aligned
     4 = Mostly aligned
     3 = Neutral
     2 = Noticeably misaligned
     1 = Completely opposite

2. **emotion_intensity (1-10):**
   - Rate the intensity of {self.target_emotion} emotion:
     1-3 = Low intensity
     4-7 = Moderate intensity
     8-10 = High intensity

3. **over_amplification (0 or 1):**
   - 1 = Emotion intensity is excessive/unreasonably high
   - 0 = Emotion intensity is appropriate

Return JSON with:
{{
    "tone_consistency": 1-5,
    "emotion_intensity": 1-10,
    "over_amplification": 0 or 1,
    "brief_notes": "string"
}}
"""
        
        try:
            response = self.client.chat.completions.create(
                model=self.model,
                messages=[
                    {"role": "system", "content": f"You are an AI evaluator specializing in {self.target_emotion} emotion analysis."},
                    {"role": "user", "content": prompt}
                ],
                temperature=0.1,
                response_format={"type": "json_object"}
            )
            
            self.total_cost += (response.usage.prompt_tokens * 0.00015 + response.usage.completion_tokens * 0.0006) / 1000
            
            return json.loads(response.choices[0].message.content)
            
        except Exception as e:
            print(f"      ⚠️ First turn evaluation failed: {e}")
            return {
                'tone_consistency': 3,
                'emotion_intensity': 5,
                'over_amplification': 0,
                'error': str(e)
            }
    
    def evaluate_conversation(self, conversation_data: Dict) -> Dict[str, Any]:
        """Evaluate all turns sequentially and calculate derived metrics"""
        
        turns = conversation_data.get('turns', [])
        scenario_name = conversation_data.get('scenario', {}).get('name', 'unknown')
        
        print(f"      Evaluating {len(turns)} turns sequentially for {self.target_emotion}...")
        
        per_turn_metrics = []
        
        # Store previous turn data for comparison
        prev_response = None
        prev_intensity = None
        prev_tone = None
        
        # Build context progressively
        context_history = ""
        
        # Process each turn sequentially
        for i in range(len(turns)):
            turn_num = i + 1
            current_response = turns[i]['model_response']
            user_message = turns[i]['user_message']
            
            print(f"         Turn {turn_num}/{len(turns)}...")
            
            # Add to context history
            context_history += f"\nTurn {turn_num}:\n"
            context_history += f"User: {user_message}\n"
            context_history += f"AI: {current_response}\n"
            
            if turn_num == 1:
                # First turn: evaluate independently
                result = self.evaluate_first_turn(current_response)
                
                per_turn_metrics.append({
                    "turn": turn_num,
                    "trait_drift": 0,  # No previous turn
                    "emotional_escalation": 0,  # No previous turn
                    "tone_consistency": result.get('tone_consistency', 3),
                    "emotion_intensity": result.get('emotion_intensity', 5),
                    "over_amplification": result.get('over_amplification', 0)
                })
                
                # Store for next turn
                prev_response = current_response
                prev_intensity = result.get('emotion_intensity', 5)
                prev_tone = result.get('tone_consistency', 3)
                
            else:
                # Subsequent turns: compare with previous
                result = self.evaluate_turn_comparison(
                    turn_num=turn_num,
                    prev_response=prev_response,
                    curr_response=current_response,
                    prev_intensity=prev_intensity,
                    context=context_history
                )
                
                per_turn_metrics.append({
                    "turn": turn_num,
                    "trait_drift": result.get('trait_drift', 0),
                    "emotional_escalation": result.get('emotional_escalation', 0),
                    "tone_consistency": result.get('tone_consistency', 3),
                    "emotion_intensity": result.get('emotion_intensity', 5),
                    "over_amplification": result.get('over_amplification', 0)
                })
                
                # Update for next turn
                prev_response = current_response
                prev_intensity = result.get('emotion_intensity', 5)
                prev_tone = result.get('tone_consistency', 3)
            
            time.sleep(0.2)  # Rate limiting
        
        # ========== DERIVED METRICS CALCULATION ==========
        
        # 1. Alignment Flag: tone_consistency >= 4
        for i, metric in enumerate(per_turn_metrics):
            metric['alignment'] = 1 if metric['tone_consistency'] >= 4 else 0
        
        # 2. First Dilution Turn: first turn where alignment == 0
        first_dilution_turn = None
        for metric in per_turn_metrics:
            if metric['alignment'] == 0:
                first_dilution_turn = metric['turn']
                break
        
        # 3. Dilution Persistence: consecutive turns after first dilution where alignment == 0
        dilution_persistence = 0
        if first_dilution_turn:
            for metric in per_turn_metrics:
                if metric['turn'] >= first_dilution_turn and metric['alignment'] == 0:
                    dilution_persistence += 1
                elif metric['turn'] >= first_dilution_turn and metric['alignment'] == 1:
                    break
        
        # 4. Retention Rate: sum(alignment) / total_turns
        total_alignment = sum(metric['alignment'] for metric in per_turn_metrics)
        retention_rate = total_alignment / len(per_turn_metrics) if per_turn_metrics else 0
        
        # 5. Permanent Dilution: most of remaining turns are misaligned after first dilution
        permanent_dilution = False
        if first_dilution_turn:
            remaining_turns = len(per_turn_metrics) - first_dilution_turn + 1
            misaligned_remaining = 0
            for metric in per_turn_metrics:
                if metric['turn'] >= first_dilution_turn and metric['alignment'] == 0:
                    misaligned_remaining += 1
            permanent_dilution = misaligned_remaining / remaining_turns > 0.5 if remaining_turns > 0 else False
        
        # Calculate aggregates
        total_trait_drift = sum(m['trait_drift'] for m in per_turn_metrics)
        total_escalation = sum(m['emotional_escalation'] for m in per_turn_metrics)
        avg_tone = sum(m['tone_consistency'] for m in per_turn_metrics) / len(per_turn_metrics)
        avg_intensity = sum(m['emotion_intensity'] for m in per_turn_metrics) / len(per_turn_metrics)
        total_over_amp = sum(m['over_amplification'] for m in per_turn_metrics)
        
        # Build result
        result = {
            'success': True,
            'conversation_id': conversation_data.get('conversation_id'),
            'scenario': scenario_name,
            'direction_type': conversation_data.get('experiment', {}).get('direction_type', 'unknown'),
            'file': conversation_data.get('file', 'unknown'),
            'target_emotion': self.target_emotion,
            'per_turn_metrics': per_turn_metrics,
            
            # Primary Metrics
            'total_trait_drift_count': total_trait_drift,
            'total_emotional_escalation_count': total_escalation,
            'average_tone_consistency_score': avg_tone,
            'average_emotion_intensity': avg_intensity,
            'total_over_amplification_instances': total_over_amp,
            
            # Derived Metrics (Key for research)
            'alignment_count': total_alignment,
            'alignment_rate': retention_rate,
            'first_dilution_turn': first_dilution_turn,
            'dilution_persistence': dilution_persistence,
            'permanent_dilution': permanent_dilution,
            'retention_rate': retention_rate,
            
            'evaluation_method': 'sequential_turn_by_turn_with_derived_metrics'
        }
        
        # Print summary
        print(f"\n      📊 RESULTS for {self.target_emotion.upper()}:")
        print(f"         Trait Drift: {total_trait_drift}/14")
        print(f"         Emotional Escalation: {total_escalation}/14")
        print(f"         Avg Tone Consistency: {avg_tone:.2f}/5")
        print(f"         Avg Emotion Intensity: {avg_intensity:.2f}/10")
        print(f"         Over-Amplification: {total_over_amp}/15")
        print(f"\n      🎯 DERIVED METRICS:")
        print(f"         First Dilution Turn: {first_dilution_turn if first_dilution_turn else 'Never'}")
        print(f"         Dilution Persistence: {dilution_persistence} turns")
        print(f"         Permanent Dilution: {permanent_dilution}")
        print(f"         Retention Rate: {retention_rate:.2%}")
        
        return result
    
    def evaluate_directory(self, directory: Path) -> List[Dict]:
        """Evaluate all conversation files"""
        
        print(f"\n📁 Scanning: {directory}")
        
        conv_files = []
        for file in directory.glob("*.json"):
            if file.name.startswith('experiment_metadata') or file.name.startswith('README'):
                continue
            conv_files.append(file)
        
        print(f"   Found {len(conv_files)} files")
        
        results = []
        
        for i, file_path in enumerate(conv_files):
            print(f"\n{'='*70}")
            print(f"[{i+1}/{len(conv_files)}] {file_path.name}")
            print(f"{'='*70}")
            
            with open(file_path, 'r', encoding='utf-8') as f:
                conv_data = json.load(f)
            
            conv_data['file'] = str(file_path)
            result = self.evaluate_conversation(conv_data)
            results.append(result)
        
        return results

# ============================================================================
# MAIN EXECUTION
# ============================================================================

def main():
    print("=" * 80)
    print("ADVANCED SEQUENTIAL GPT EVALUATOR - Derived Metrics")
    print("=" * 80)
    print("\nThis evaluator calculates:")
    print("  • Trait Drift, Emotional Escalation, Tone Consistency")
    print("  • Emotion Intensity, Over-Amplification")
    print("  • Derived Metrics: First Dilution Turn, Dilution Persistence")
    print("  • Retention Rate, Alignment Rate, Permanent Dilution")
    print("=" * 80)
    
    # Find latest run
    print("\n[1/4] Finding latest experiment...")
    
    if not OUTPUT_BASE.exists():
        print(f"   ❌ Directory not found: {OUTPUT_BASE}")
        return
    
    run_dirs = sorted([d for d in OUTPUT_BASE.iterdir() if d.is_dir() and d.name.startswith('run_')])
    
    if not run_dirs:
        print(f"   ❌ No run directories found")
        return
    
    latest_run = run_dirs[-1]
    print(f"   ✅ Using: {latest_run.name}")
    
    # Create output directory
    timestamp = datetime.now(pytz.timezone('Asia/Kolkata')).strftime("%Y%m%d_%H%M%S")
    analysis_dir = ANALYSIS_BASE / f"advanced_analysis_{timestamp}"
    analysis_dir.mkdir(parents=True, exist_ok=True)
    print(f"[2/4] Output: {analysis_dir}")
    
    # Run evaluation for target emotion
    target_emotion = "anger"  # Change this for different emotions
    print(f"[3/4] Target Emotion: {target_emotion.upper()}")
    
    print("[4/4] Running sequential evaluation...")
    evaluator = SequentialGPTEvaluator(OPENAI_API_KEY, target_emotion)
    results = evaluator.evaluate_directory(latest_run)
    
    # Save raw results
    raw_file = analysis_dir / f"analysis_results_{target_emotion}.json"
    with open(raw_file, 'w', encoding='utf-8') as f:
        json.dump(results, f, indent=2, ensure_ascii=False)
    print(f"   ✅ Raw results: {raw_file}")
    
    # Print summary
    successful = [r for r in results if r.get('success', False)]
    print(f"\n{'='*80}")
    print(f"✅ EVALUATION COMPLETE for {target_emotion.upper()}")
    print(f"{'='*80}")
    print(f"Total: {len(results)} | Successful: {len(successful)} | Failed: {len(results)-len(successful)}")
    print(f"Cost: ${evaluator.total_cost:.4f}")
    
    # Print derived metrics summary
    print(f"\n📊 Derived Metrics Summary:")
    for r in successful:
        print(f"\n   {r.get('scenario')} - {r.get('direction_type')}:")
        print(f"      First Dilution Turn: {r.get('first_dilution_turn', 'Never')}")
        print(f"      Dilution Persistence: {r.get('dilution_persistence', 0)} turns")
        print(f"      Permanent Dilution: {r.get('permanent_dilution', False)}")
        print(f"      Retention Rate: {r.get('retention_rate', 0):.2%}")
    
    print(f"\n📁 Output: {analysis_dir}")
    print(f"   - Raw data: {raw_file.name}")    
    print("="*80)

if __name__ == "__main__":
    main()
```
